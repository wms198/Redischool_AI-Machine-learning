```python
# Titanic Dataset Python Basics Review

# Libraries we'll use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For pretty plots
%matplotlib inline

# Set seaborn style
sns.set(style='whitegrid')
```

```python
# Load the Titanic dataset
# Make sure the CSV is in the same directory or update the path
df = pd.read_csv("titanic-datasets/train.csv")

# Preview the first few rows
df.head(1000)

```

```python
plt.figure(figsize=(7,3))
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
custom_palette = {'male': 'lightblue', 'female': 'pink'}
sns.pairplot(df, hue='Sex', palette=custom_palette)
plt.show()
```
